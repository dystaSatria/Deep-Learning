```python
iris.drop('Id', axis=1, inplace=True)
```

```python
iris.head()
```

```python
target_species = {k:v for k, v in zip(iris['Species'].unique(),
                                      list(range(len(iris['Species'].unique()))))}
```

```python
target_species
```

```python
iris['Species'] = iris['Species'].map(target_species)
```

```python
iris.head()
```

```python
x = iris.iloc[:,:4]
```

```python
y = iris.iloc[:,4]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# from sklearn.metrics import accuracy_score
# Decision Tree

from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state = 13)
dt_model.fit(x_train,y_train) #modeli eğitmek
dt_predict = dt_model.predict(x_test)



```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate the metrics for the Decision Tree classifier
dt_accuracy = accuracy_score(y_test, dt_predict)
dt_precision = precision_score(y_test, dt_predict, average='weighted')
dt_recall = recall_score(y_test, dt_predict, average='weighted')
dt_f1 = f1_score(y_test, dt_predict, average='weighted')

print("Decision Tree Accuracy Score:", dt_accuracy)
print("Decision Tree Precision Score:", dt_precision)
print("Decision Tree Recall Score:", dt_recall)
print("Decision Tree F1 Score:", dt_f1)
```

```python
from sklearn.metrics import classification_report

labels = ['Iris-setosa','Iris-virginica','Iris-versicolor']

rf_report = classification_report(y_test,dt_predict,target_names = labels)
print(rf_report)
```
