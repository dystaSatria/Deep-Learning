```python
iris.drop('Id', axis=1, inplace=True)
```

```python
iris.head()
```

```python
target_species = {k:v for k, v in zip(iris['Species'].unique(),
                                      list(range(len(iris['Species'].unique()))))}
```

```python
target_species
```

```python
iris['Species'] = iris['Species'].map(target_species)
```

```python
iris.head()
```

```python
x = iris.iloc[:,:4]
```

```python
y = iris.iloc[:,4]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC

# Membuat model SVM dengan kernel linear
svm_model = SVC(kernel='linear', random_state=0)
svm_model.fit(x_train, y_train)  # Melatih model SVM
svml_predict = svm_model.predict(x_test)  # Melakukan prediksi


```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


print("SVM Line accuracy Score:", accuracy_score(y_test, svml_predict))
print("SVM Line precision Score:", precision_score(y_test, svml_predict, average='weighted'))
print("SVM Line recall Score:", recall_score(y_test, svml_predict, average='weighted'))
print("SVM Line f1 Score:", f1_score(y_test, svml_predict, average='weighted'))




```

```python
from sklearn.metrics import classification_report

labels = ['Iris-setosa','Iris-virginica','Iris-versicolor']

svml_report = classification_report(y_test,svml_predict,target_names = labels)
print(svml_report)
```
