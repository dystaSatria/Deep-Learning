```
# Image Classification using Transfer Learning and SVM
# -------------------------------------------------
# This code implements a pipeline for image classification:
# 1. Uses pretrained DenseNet models (121, 169, 201) for feature extraction
# 2. Extracts features from images using these models
# 3. Trains SVM classifiers with different kernels on the extracted features
# 4. Evaluates performance using multiple metrics
# 5. Visualizes results with confusion matrices and ROC curves

# Import necessary libraries
# -------------------------------------------------

# Basic data manipulation and visualization
import numpy as np              # For numerical operations
import pandas as pd             # For data manipulation and analysis
import matplotlib.pyplot as plt # For plotting
import seaborn as sns           # For enhanced visualizations
import os                       # For file operations
import time                     # For timing operations
from tqdm import tqdm           # For progress bars
import warnings                 # For suppressing warnings
warnings.filterwarnings('ignore')

# TensorFlow and Keras for deep learning
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121, DenseNet169, DenseNet201  # The transfer learning models
from tensorflow.keras.preprocessing.image import ImageDataGenerator              # For loading and preprocessing images
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense                # For model architecture
from tensorflow.keras.models import Model                                        # For creating models

# Scikit-learn for machine learning and evaluation
from sklearn.model_selection import train_test_split  # For splitting data
from sklearn.preprocessing import StandardScaler      # For feature scaling
from sklearn.svm import SVC                           # Support Vector Machine classifier
# Metrics for evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score
from sklearn.metrics import matthews_corrcoef, cohen_kappa_score
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Mount Google Drive (for Colab)
# This gives access to your Google Drive files
from google.colab import drive
drive.mount('/content/drive')

# Configuration parameters
# -------------------------------------------------
# This class centralizes all configuration parameters
class Config:
    # Dataset parameters
    DATA_DIR = '/content/drive/MyDrive/Combined Dataset'  # Path to your dataset - UPDATE THIS!
    IMG_SIZE = 224       # Input image size (224x224 pixels is standard for these models)
    BATCH_SIZE = 32      # Number of images processed in each batch

    # Transfer learning models to use
    MODELS = {
        'DenseNet121': DenseNet121,  # Lighter model, fewer parameters
        'DenseNet169': DenseNet169,  # Medium model
        'DenseNet201': DenseNet201   # Deeper model, more parameters
    }

    # SVM kernel types
    SVM_KERNELS = ['linear', 'rbf', 'sigmoid', 'poly']  # Different kernel functions for SVM

    # Class names (will be updated automatically based on folder names)
    CLASS_NAMES = ['class1', 'class2', 'class3']  # Placeholder - will be replaced
    NUM_CLASSES = len(CLASS_NAMES)

config = Config()

# Data generators for loading and preprocessing images
# -------------------------------------------------
def create_data_generators():
    """
    Creates data generators to load and preprocess images from directories.
    Returns train and validation generators.

    The dataset should be structured as:
    - DATA_DIR/
        - class1/
            - image1.jpg
            - image2.jpg
            ...
        - class2/
            ...
    """
    # Data augmentation for training to improve model generalization
    # These transformations create variations of the original images
    train_datagen = ImageDataGenerator(
        rescale=1./255,                # Normalize pixel values to [0,1]
        rotation_range=20,             # Randomly rotate images by up to 20 degrees
        width_shift_range=0.2,         # Randomly shift images horizontally
        height_shift_range=0.2,        # Randomly shift images vertically
        shear_range=0.2,               # Shear transformations
        zoom_range=0.2,                # Zoom transformations
        horizontal_flip=True,          # Randomly flip images horizontally
        validation_split=0.2           # Use 20% of data for validation
    )

    # For validation, we only rescale (normalize) the images
    test_datagen = ImageDataGenerator(rescale=1./255)

    # Create data generator for training data
    train_generator = train_datagen.flow_from_directory(
        config.DATA_DIR,                        # Input directory
        target_size=(config.IMG_SIZE, config.IMG_SIZE),  # Resize images
        batch_size=config.BATCH_SIZE,           # Batch size
        class_mode='categorical',               # For multi-class classification
        subset='training',                      # Use training subset
        shuffle=True                            # Shuffle the data
    )

    # Create data generator for validation data
    validation_generator = train_datagen.flow_from_directory(
        config.DATA_DIR,
        target_size=(config.IMG_SIZE, config.IMG_SIZE),
        batch_size=config.BATCH_SIZE,
        class_mode='categorical',
        subset='validation',                    # Use validation subset
        shuffle=False                           # Don't shuffle validation data
    )

    # Update class names based on the directory structure
    # This automatically detects your classes from folder names
    config.CLASS_NAMES = list(train_generator.class_indices.keys())
    config.NUM_CLASSES = len(config.CLASS_NAMES)

    return train_generator, validation_generator

# Function to create the transfer learning model for feature extraction
# -------------------------------------------------
def create_feature_extractor(model_name):
    """
    Creates a feature extractor model based on the specified DenseNet variant.
    The model is pre-trained on ImageNet and has its top layer removed.

    Args:
        model_name: Name of the model ('DenseNet121', 'DenseNet169', or 'DenseNet201')

    Returns:
        A Keras model that outputs feature vectors
    """
    # Get the model class from our config
    base_model_class = config.MODELS[model_name]

    # Create the base model with ImageNet weights, without the classification layer
    base_model = base_model_class(
        weights='imagenet',                     # Use pre-trained weights from ImageNet
        include_top=False,                      # Remove the top classification layer
        input_shape=(config.IMG_SIZE, config.IMG_SIZE, 3)  # Input shape (RGB images)
    )

    # Freeze the base model layers (no training)
    # This is important for transfer learning - we use the pre-trained weights as is
    base_model.trainable = False

    # Add a global spatial average pooling layer to convert the feature maps to feature vectors
    x = base_model.output
    x = GlobalAveragePooling2D()(x)

    # Create the final model for feature extraction
    model = Model(inputs=base_model.input, outputs=x)

    return model

# Function to extract features using the transfer learning model
# -------------------------------------------------
def extract_features(model, data_generator):
    """
    Extracts features from images using the given model.

    Args:
        model: The feature extraction model
        data_generator: Generator that provides batches of images

    Returns:
        features: Extracted features (numpy array)
        labels: Corresponding labels (numpy array)
    """
    features = []
    labels = []

    # Get the number of batches
    steps = len(data_generator)

    # Reset the generator (important to start from the beginning)
    data_generator.reset()

    # Extract features batch by batch
    for i in tqdm(range(steps), desc=f"Extracting features"):
        # Get a batch of images and their labels
        x, y = next(data_generator)

        # Predict (extract) features using the model
        batch_features = model.predict(x, verbose=0)

        # Store features and labels
        features.append(batch_features)
        labels.append(y)

    # Concatenate all batches
    features = np.vstack(features)
    labels = np.vstack(labels)

    # Convert one-hot encoded labels back to class indices
    # (SVM expects class labels as integers, not one-hot encoded vectors)
    labels = np.argmax(labels, axis=1)

    return features, labels

# Function to train SVM with different kernels and evaluate performance
# -------------------------------------------------
def train_and_evaluate_svm(X_train, y_train, X_test, y_test, model_name):
    """
    Trains SVM classifiers with different kernels and evaluates their performance.

    Args:
        X_train: Training features
        y_train: Training labels
        X_test: Testing features
        y_test: Testing labels
        model_name: Name of the feature extraction model

    Returns:
        List of dictionaries containing results for each kernel
    """
    results = []

    # Scale the features (important for SVM performance)
    # StandardScaler standardizes features by removing the mean and scaling to unit variance
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)  # Fit on training data and transform
    X_test_scaled = scaler.transform(X_test)        # Transform test data using same parameters

    # Loop through different SVM kernels
    for kernel in config.SVM_KERNELS:
        print(f"\nTraining SVM with {kernel} kernel for {model_name}...")

        # Create and train the SVM model
        # kernel: The type of kernel function
        # probability=True: Enable probability estimates (needed for ROC curve)
        # random_state=42: For reproducibility
        svm = SVC(kernel=kernel, probability=True, random_state=42)
        svm.fit(X_train_scaled, y_train)

        # Make predictions on test data
        y_pred = svm.predict(X_test_scaled)

        # Calculate probabilities for ROC curve
        y_prob = svm.predict_proba(X_test_scaled)

        # Calculate evaluation metrics
        # Note: 'weighted' average accounts for class imbalance
        accuracy = accuracy_score(y_test, y_pred)  # Overall accuracy
        precision = precision_score(y_test, y_pred, average='weighted')  # Precision: TP/(TP+FP)
        recall = recall_score(y_test, y_pred, average='weighted')        # Recall: TP/(TP+FN)
        f1 = f1_score(y_test, y_pred, average='weighted')                # F1: 2*(Precision*Recall)/(Precision+Recall)
        f2 = fbeta_score(y_test, y_pred, beta=2, average='weighted')     # F2: Emphasizes recall over precision
        mcc = matthews_corrcoef(y_test, y_pred)                          # MCC: Correlation coefficient between actual and predicted
        kappa = cohen_kappa_score(y_test, y_pred)                        # Kappa: Agreement level compared to random chance

        # Print results
        print(f"Accuracy: {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1 Score: {f1:.4f}")
        print(f"F2 Score: {f2:.4f}")
        print(f"MCC: {mcc:.4f}")
        print(f"Kappa: {kappa:.4f}")

        # Create confusion matrix
        # Shows how many instances of each class were correctly/incorrectly classified
        cm = confusion_matrix(y_test, y_pred)

        # Plot confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=config.CLASS_NAMES, yticklabels=config.CLASS_NAMES)
        plt.title(f'Confusion Matrix - {model_name} with {kernel} kernel')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.show()

        # Plot ROC curve (for multi-class, we use One-vs-Rest approach)
        plt.figure(figsize=(10, 8))

        # Calculate ROC curve and ROC area for each class
        fpr = dict()  # False positive rate
        tpr = dict()  # True positive rate
        roc_auc = dict()  # Area under ROC curve

        # Calculate ROC curve for each class
        for i in range(config.NUM_CLASSES):
            # Convert to binary classification (one class vs. rest)
            fpr[i], tpr[i], _ = roc_curve((y_test == i).astype(int), y_prob[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            # Plot the curve
            plt.plot(fpr[i], tpr[i], lw=2,
                     label=f'ROC curve of class {config.CLASS_NAMES[i]} (area = {roc_auc[i]:.2f})')

        # Calculate micro-average ROC curve and ROC area
        # This combines all classes into one ROC curve
        y_test_bin = tf.keras.utils.to_categorical(y_test, num_classes=config.NUM_CLASSES)
        fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_prob.ravel())
        roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

        # Plot the micro-average ROC curve
        plt.plot(fpr["micro"], tpr["micro"],
                 label=f'micro-average ROC curve (area = {roc_auc["micro"]:.2f})',
                 color='deeppink', linestyle=':', linewidth=4)

        # Plot the diagonal (random classifier line)
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve - {model_name} with {kernel} kernel')
        plt.legend(loc="lower right")
        plt.show()

        # Store results in a dictionary
        result = {
            'model': model_name,
            'kernel': kernel,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'f2': f2,
            'mcc': mcc,
            'kappa': kappa,
            'roc_auc': roc_auc["micro"],
            'confusion_matrix': cm
        }

        # Add to results list
        results.append(result)

    return results

# Main execution function
# -------------------------------------------------
def main():
    """
    Main function to execute the entire pipeline:
    1. Load data
    2. Extract features using different models
    3. Train and evaluate SVM with different kernels
    4. Compare results
    """
    # Create data generators
    train_generator, validation_generator = create_data_generators()

    # Print information about the dataset
    print(f"Class names: {config.CLASS_NAMES}")
    print(f"Number of classes: {config.NUM_CLASSES}")
    print(f"Training samples: {train_generator.samples}")
    print(f"Validation samples: {validation_generator.samples}")

    # Store all results
    all_results = []

    # Loop through each model (DenseNet121, DenseNet169, DenseNet201)
    for model_name in config.MODELS.keys():
        print(f"\n\n===== Processing {model_name} =====")

        # Create the feature extractor
        model = create_feature_extractor(model_name)

        # Extract features for training and validation data
        print(f"Extracting features using {model_name}...")
        train_features, train_labels = extract_features(model, train_generator)
        val_features, val_labels = extract_features(model, validation_generator)

        print(f"Train features shape: {train_features.shape}")
        print(f"Validation features shape: {val_features.shape}")

        # Train SVM with different kernels and evaluate
        results = train_and_evaluate_svm(train_features, train_labels, val_features, val_labels, model_name)
        all_results.extend(results)

    # Create a DataFrame with all results for easy comparison
    results_df = pd.DataFrame([
        {k: v for k, v in result.items() if k != 'confusion_matrix'}  # Exclude confusion matrix
        for result in all_results
    ])

    # Display the results
    print("\n\n===== Summary of Results =====")
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', 1000)
    print(results_df)

    # Save results to CSV
    results_df.to_csv('model_comparison_results.csv', index=False)
    print("Results saved to 'model_comparison_results.csv'")

    # Plot comparison of models and kernels
    plt.figure(figsize=(15, 10))

    # Metrics to compare
    metrics = ['accuracy', 'precision', 'recall', 'f1', 'f2', 'mcc', 'kappa', 'roc_auc']

    # Create subplots for each metric
    for i, metric in enumerate(metrics, 1):
        plt.subplot(2, 4, i)  # 2 rows, 4 columns
        # Create a bar plot comparing models and kernels for this metric
        sns.barplot(x='kernel', y=metric, hue='model', data=results_df)
        plt.title(f'{metric.upper()} Comparison')
        plt.xticks(rotation=45)
        if i >= 5:  # Only add xlabel to bottom plots
            plt.xlabel('SVM Kernel')
        plt.ylabel(metric.upper())

    plt.tight_layout()
    plt.savefig('metrics_comparison.png')
    plt.show()
    print("Comparison plot saved to 'metrics_comparison.png'")

# Execute the main function when the script is run
if __name__ == "__main__":
    main()
```
